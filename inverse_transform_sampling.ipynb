### Метод обратного преобразования

```python
import random
import matplotlib.pyplot as plt
import math
```

#### Создаем выборку из непрерывного равномерного распределения

```python
uniform_lst = [random.random() for i in range(100000)] ## Генерируем 100К значений из непрерывного равномерного распределния 
```

#### Строим плотность вероятности

```python
plt.hist(uniform_lst, color = 'blue', edgecolor = 'black')
plt.show()
```

#### Создаем выборку экспоненциального распределения при помощи обратного преобразования

```python
expo_lst = list(map((lambda x, halflife =1: -1/halflife * math.log(1-x)), uniform_lst ))
```

#### Строим плотность вероятности

```python
plt.hist(expo_lst, color = 'blue', edgecolor = 'black', bins = 100)
plt.show()
```

#### Создаем выборку распределения Пуасона при помощи обратного преобразования

```python
poisson_lst = list(map(lambda x, k=1, xm=1: xm / ((1 -x)**1/k), uniform_lst))
```

#### Строим плотность вероятности

```python
plt.hist(poisson_lst, color = 'blue', edgecolor = 'black', bins = 100000)
plt.xlim(0, 25)
plt.show()
```
